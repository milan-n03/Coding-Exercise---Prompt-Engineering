**Exercise 3: Self-Reflection Prompt for Improving Output**

**Tool Used:** ChatGPT (GPT-4) + Google Colab

**Step 1:** Original Summary

**Prompt:**

You are an academic writing assistant summarizing articles for a college-level audience.

Summarize the following article in a **single paragraph** of exactly **150–180 words**.

Focus on:
- The article's main argument
- Key ethical challenges discussed (e.g., bias, privacy, transparency)
- Broader societal implications

Constraints:
- Write in plain academic English — avoid business jargon
- Do NOT use bullet points or subheadings
- Do NOT include personal opinion
- Output must be one paragraph only

Article: https://professional.dce.harvard.edu/blog/ethics-in-ai-why-it-matters/

**Output:** The article stresses the growing importance of ethical considerations as artificial intelligence becomes deeply embedded in business operations and decision-making. As companies adopt AI tools for efficiency and competitive advantage, leaders must understand not just how AI works, but the ethical implications of its use. Ethical literacy allows executives to anticipate risks, protect their organizations from legal and reputational harm, and balance opportunities with responsibilities. Central ethical challenges include data privacy and security, ensuring that sensitive information stays protected; bias in AI systems, which can emerge from training data, algorithms, or developers' assumptions; and model transparency, where leaders need to understand and explain how AI reaches decisions rather than treat systems as "black boxes."

Beyond these challenges, the article highlights the broader societal impact of AI, such as its potential to reshape the workforce and environmental considerations tied to energy-intensive AI data centers. Because ethical issues in AI are complex and evolving, there is no single roadmap for resolving them; instead, leaders must stay informed, creatively navigate ethical dilemmas, and build governance frameworks that uphold fairness, accountability, and trust. Responsible AI use can offer competitive advantage and help organizations avoid pitfalls that arise when ethics are overlooked.

**Step 2:** Ask ChatGPT to Critique Itself

**Prompt:**

You will critique your own summary using the following criteria:

1. Accuracy – Does it correctly represent the key ideas? Identify missing or distorted points.
2. Clarity – Is the writing clear, concise, and logically structured?
3. Audience – Is it appropriate for a college-level reader?
4. Length – The summary must be between 150–180 words.
5. Formatting – The summary must be one paragraph only.

Instructions:
- First, list weaknesses under each criterion.
- Then provide a fully revised improved summary.
- Do NOT repeat the original summary.
- Ensure the revised version satisfies all requirements.

**Output:**

1) Accuracy — weaknesses

My summary left out major sections of the article, especially AI governance, the regulatory landscape (U.S. vs. EU), and the article’s employment/UBI discussion.

It flattened specifics (e.g., the article’s point that bias can come from programmers, the algorithm, or training data).

It implied broad “frameworks” without reflecting the article’s emphasis that there is no one-size-fits-all, step-by-step process because AI use cases vary.

2) Clarity — weaknesses

A few sentences were overpacked (many ideas crammed into one line), which reduced readability.

The structure could be more signposted (moving from business risks → technical ethics issues → societal/governance issues more cleanly).

3) Audience — weaknesses

Mostly appropriate, but a couple phrases (e.g., “digital orchestration”-style wording) were more business-jargon than needed for a general college audience.

4) Length — weaknesses

It missed the target (I wrote ~200 words, but the requirement is 150–180).

5) Formatting — weaknesses

It used multiple paragraphs instead of one single paragraph.




**Step 3:** Revised Summary

**Output:**

The article argues that as companies embed AI into everyday operations, leaders need “AI ethics literacy” to anticipate harms and protect organizations from lawsuits and reputational damage while making balanced decisions about AI’s opportunities and risks. It highlights three core problem areas: safeguarding sensitive and personally identifying data from new AI-enabled attack paths; recognizing and mitigating bias that can originate in programmers, algorithms, or training data; and demanding transparency so humans can explain AI decisions in high-stakes domains like finance, healthcare, and education. The piece also emphasizes broader societal effects, including the strain AI data centers can place on resources and the possibility of major white-collar job disruption that may outpace retraining, raising ideas such as government support or even universal basic income. Finally, it notes that governance is emerging through international principles (OECD) and UN advisory efforts, while regulation remains fluid—contrasting U.S. debates with the EU’s risk-based AI Act—and concludes that firms that operationalize ethical “gut checks” around privacy, fairness, and explainability can build trust and gain competitive advantage.
